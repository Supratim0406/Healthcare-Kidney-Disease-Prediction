```python
data = df.sample(frac=0.8, random_state=42).reset_index(drop=True)
data_unseen = df.drop(data.index).reset_index(drop=True)
```

## Automation method by using Pycaret Library

```python
!pip install pycaret
```

```python
from pycaret.classification import *
```

```python
exp_classification = setup(data = data, target = 'Classification', session_id=123)
```

```python
compare_models()
```

```python
et = create_model('et')
```

```python
## Fine tunning
tuned_et = tune_model(et)
```

```python
## Evalute the model
evaluate_model(et)
```

```python
## Predict the Unseen dataset
unseen_prediction =  predict_model(et, data=data_unseen)
unseen_prediction.head()
```

```python
## save model
save_model(et, 'et_model')
```
